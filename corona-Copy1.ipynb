```python
import matplotlib.pyplot as plt
import numpy
import seaborn as sbn
```

```python
start = 95
start_threat = 40
dp_surv = []
fp_surv = []
ip_surv = []

dp_threat = []
fp_threat = []
ip_threat = []

for i in range(0,10):
    dp_surv.append(start - i * 5 + 6.6)
    fp_surv.append(start - i * 5)
    ip_surv.append(start - i * 5)
    
    dp_threat.append(start_threat + i * 5)
    fp_threat.append(start_threat + 7 + i * 9)
    ip_start_threat = start_threat + i * 5
    ip_threat.append(ip_start_threat * (1+.0018*ip_start_threat))
```

```python
dp, = plt.plot(dp_threat, dp_surv, label="Deep Prot")
fp, = plt.plot(fp_threat, fp_surv, label="Fury Prot")
ip, = plt.plot(ip_threat, ip_surv, label="Impale Prot")
plt.xlabel("Threat")
plt.ylabel("Survivability")
plt.legend(handles=[dp, fp, ip])
plt.grid()
plt.axis([30, 150, 40, 110])
```
